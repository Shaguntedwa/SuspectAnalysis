```python
import pandas as pd
df=pd.read_csv('/content/school_data.csv')
print("Here's a sneak-peek at the data.")
print(df.head())
print("\nInfo:")
print(df.info())
print('\n Student with more than 10 choclate wrappers:')
print(df[df['ChocolateWrappers']>10])
print("\n Top 3 heaviest backpacks:")
print(df.sort_values('BackpackWeight(kg)',ascending=False).head(3))
print("\n Student with more than 20 minutes in library:")
print(df[df['TimeInLibrary(min)']<20])
print("\n Average detentions per house:")
print(df.groupby('House')['Detentions'].mean())
def is_suspect(row):
  conditions_met=0
  if row['BackpackWeight(kg)']>6.5:
    conditions_met+=1
  if row['ChocolateWrappers']>10:
    conditions_met+=1
  if row['TimeInLibrary(min)']<30:
    conditions_met+=1
  return conditions_met>=2
df['Suspect']=df.apply(is_suspect,axis=1)
suspects=df[df["Suspect"]==True]
print("\n Final Suspect List:")
print(suspects[['Name','House','BackpackWeight(kg)','ChocolateWrappers','TimeInLibrary(min)']])
suspects.to_csv("suspect_list.csv",index="False")
print("\n Suspect list saved to 'susepct_list.csv")
```
